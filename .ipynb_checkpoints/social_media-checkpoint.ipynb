```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib notebook
plt.style.use('ggplot')
```

```python
raw_data = pd.read_csv('survey.csv')
```

```python
raw_data.head()
```

```python
data = raw_data
data = data.drop(['StartDate', 'EndDate', 'Status', 'IPAddress', 'Duration (in seconds)', 
                    'RecordedDate', 'ResponseId', 'Progress', 'Finished', 'RecipientLastName', 
                    'RecipientFirstName', 'RecipientEmail', 'ExternalReference', 'LocationLatitude', 
                    'DistributionChannel', 'UserLanguage', 'LocationLongitude'], axis=1)
data.head()
```

```python
frequency = list(data['Q3'][2:])

freqdict = {'Never': 0, 'Less often': 0, 'Once every few weeks': 0, 
            'About once a week': 0, 'Several times a week': 0, 'About once a day': 0,
            'Several times a day': 0}

for row in frequency:
    if row == 'Never':
        freqdict['Never'] += 1
    elif row == 'Less often':
        freqdict['Less often'] += 1
    elif row == 'Once every few weeks':
        freqdict['Once every few weeks'] += 1
    elif row == 'About once a week':
        freqdict['About once a week'] += 1
    elif row == 'Several times a week':
        freqdict['Several times a week'] += 1
    elif row == 'Several times a day':
        freqdict['Several times a day'] += 1
    elif row == 'About once a day':
        freqdict['About once a day'] += 1



print(list(freqdict.keys()))
total_num = sum(list(freqdict.values()))
```

```python
#plt.figure()
plt.figure(figsize = (8,6))
plt.xlabel('Fraction of Respondents')
# Set the y axis label of the current axis.
#plt.ylabel('Post Frequency')
# Set a title of the current axes.
plt.title('Posting Frequency')

plt.barh(list(freqdict.keys()), list(freqdict.values()))
#plt.savefig('freq.png', dpi=300, bbox_inches='tight')
```

```python
apps = list(data['Q1'][2:])
apps = ','.join(apps)
apps = apps.split(',')
appsdict = {'Facebook': 0, 'Snapchat': 0, 'Instagram': 0, 
            'Twitter': 0, 'TikTok': 0, 'LinkedIn': 0, 'GroupMe': 0, 
            'Messenger': 0, 'Other': 0}

for e in range(len(apps)):
    if apps[e] == 'Facebook':
        appsdict['Facebook'] += 1
    elif apps[e] == 'Facebook Messenger':
        appsdict['Messenger'] += 1
    elif apps[e] == 'Snapchat':
        appsdict['Snapchat'] += 1
    elif apps[e] == 'Instagram':
        appsdict['Instagram'] += 1
    elif apps[e] == 'Twitter':
        appsdict['Twitter'] += 1
    elif apps[e] == 'TikTok':
        appsdict['TikTok'] += 1
    elif apps[e] == 'LinkedIn':
        appsdict['LinkedIn'] += 1
    elif apps[e] == 'GroupMe':
        appsdict['GroupMe'] += 1
    else:
        appsdict['Other'] += 1
    
print(appsdict)
```

```python
plt.figure()
#plt.figure(figsize = (12,6))
plt.xlabel('Social Media App')
# Set the y axis label of the current axis.
plt.ylabel('Number of Respondents')
# Set a title of the current axes.
plt.title('Popularity of Selected Social Media Apps at Harvard')

plt.bar(appsdict.keys(), appsdict.values())
#plt.savefig('apps.png', dpi=300, bbox_inches='tight')
```

```python
popularity_series = data['Q24_1'][2:]

popdict = {'1': 0, '2': 0, '3': 0, '4': 0}

for row in popularity_series:
    if row == '1':
        popdict['1'] += 1
    elif row == '2':
        popdict['2'] += 1
    elif row == '3':
        popdict['3'] += 1
    elif row == '4':
        popdict['4'] += 1
print(popdict)
```

```python
plt.figure()
plt.xlabel('Popularity Score')
# Set the y axis label of the current axis.
plt.ylabel('Number of Respondents')
# Set a title of the current axes.
plt.title('Self-Reported Social Media Popularity Score')

plt.bar(popdict.keys(), popdict.values())
#plt.savefig('popularity.png', dpi=300, bbox_inches='tight')
```

```python
ethnicity_series = data['Q14'][2:]

ethnicity_dict = {'White': 0, 'Black': 0, 'Hispanic': 0, 'Asian': 0, 'Other': 0}

for row in ethnicity_series:
    if row == 'Hispanic or Latinx':
        ethnicity_dict['Hispanic'] += 1
    elif row == 'White':
        ethnicity_dict['White'] += 1
    elif row == 'Asian or Asian American':
        ethnicity_dict['Asian'] += 1
    elif row == 'Black or African American':
        ethnicity_dict['Black'] += 1
    else:
        ethnicity_dict['Other'] += 1
print(ethnicity_dict)
```

```python
plt.figure()
plt.xlabel('Ethnicity')
# Set the y axis label of the current axis.
plt.ylabel('Number of Respondents')
# Set a title of the current axes.
plt.title('Respondent Demographics')

plt.bar(ethnicity_dict.keys(), ethnicity_dict.values())
#plt.savefig('ethnicity.png', dpi=300, bbox_inches='tight')
```

```python
gender_series = data['Q15'][2:]

gender_dict = {'Male': 0, 'Female': 0, 'Non-binary': 0}

for row in gender_series:
    if row == 'Male':
        gender_dict['Male'] += 1
    elif row == 'Female':
        gender_dict['Female'] += 1
    else:
        gender_dict['Non-binary'] += 1
print(gender_dict)
```

```python
plt.figure()
plt.xlabel('Gender')
# Set the y axis label of the current axis.
plt.ylabel('Number of Respondents')
# Set a title of the current axes.
plt.title('Gender of Respondents')

plt.bar(gender_dict.keys(), gender_dict.values())
#plt.savefig('gender.png', dpi=300, bbox_inches='tight')
```

```python
class_series = data['Q20'][2:]

class_dict = {'2024': 0, '2023': 0, '2022': 0, '2021': 0}

for row in class_series:
    if row == '2021':
        class_dict['2021'] += 1
    elif row == '2022':
        class_dict['2022'] += 1
    elif row == '2023':
        class_dict['2023'] += 1
    elif row == '2024':
        class_dict['2024'] += 1
print(class_dict)
```

```python
plt.figure()
plt.xlabel('Social Class Year')
# Set the y axis label of the current axis.
plt.ylabel('Number of Respondents')
# Set a title of the current axes.
plt.title('Social Class Year of Survey Respondents')

plt.bar(class_dict.keys(), class_dict.values())
#plt.savefig('class.png', dpi=300, bbox_inches='tight')
```
